```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file into a DataFrame
df = pd.read_excel(r"C:/Users/SRI/Downloads/data.xlsx")

# Univariate Analysis for numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_cols:
    plt.figure(figsize=(10, 5))
    
    # Probability Density Function (PDF)
    sns.histplot(df[col], kde=True)
    plt.title(f"Probability Density Function (PDF) of {col}")
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.show()
    
    # Boxplot
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=df, x=col)
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
    
    # Find outliers
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]
    print(f"Outliers in {col}:", outliers)
    
    # Summary statistics
    print(f"\nSummary statistics of {col}:")
    print(df[col].describe())
    print("\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_excel(r"C:/Users/SRI/Downloads/data.xlsx")
# Boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='ID', y='Salary')
plt.title("Boxplot for Categorical vs Numerical Columns")
plt.xlabel("Categorical Column")
plt.ylabel("Numerical Column")
plt.show()

# Bar plot
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='ID', y='Salary')
plt.title("Bar Plot for Categorical vs Numerical Columns")
plt.xlabel("Categorical Column")
plt.ylabel("Numerical Column")
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency


# Create a contingency table of gender vs specialization
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Perform chi-square test for independence
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p_value)
```
